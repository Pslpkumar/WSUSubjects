```python
from PIL import Image
import glob
import numpy as np
import pandas as pd
```

```python
x_test=pd.DataFrame([])
images = glob.glob("Spoof_data/Testing Biometrika Spoof/Testing Biometrika Spoof/spoof/*.png")
for image in images:
    with open(image, 'rb') as file:
        img = Image.open(file)
        resize = (25, 25)
        img = img.resize(resize)
        arr = np.array(img)
        vector = arr.ravel()
        vector=[vector]
        image = pd.DataFrame(vector)
        x_test=x_test.append(image,ignore_index = True)
```

```python
x_test.loc[:,'label'] = 0
```
