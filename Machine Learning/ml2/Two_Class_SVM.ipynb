```python
import cv2
import glob
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
test_img=pd.DataFrame([])
for img in glob.glob("Spoof_data/Testing Biometrika Spoof/Testing Biometrika Spoof/spoof/*.png"):
    cv_img = cv2.imread(img)
    newsize = (30, 30)
    img = cv2.resize(cv_img,newsize,interpolation = cv2.INTER_AREA)
    arr = np.array(img)
    one_vector = arr.ravel()
    one_vector=[one_vector]
    image = pd.DataFrame(one_vector)
    test_img=test_img.append(image,ignore_index = True)
```

```python
test_img.head()
```

```python
test_img['label'] = np.zeros(len(test_img.index),dtype=int)
```

```python
test_img.head()
```

```python
test_img_live=pd.DataFrame([])
for img in glob.glob("Spoof_data/Testing Biometrika Live/live/*.png"):
    cv_img = cv2.imread(img)
    newsize = (30, 30)
    img = cv2.resize(cv_img,newsize,interpolation = cv2.INTER_AREA)
    arr = np.array(img)
    one_vector = arr.ravel()
    one_vector=[one_vector]
    image = pd.DataFrame(one_vector)
    test_img_live=test_img_live.append(image,ignore_index = True)
```

```python
test_img_live.head()
```

```python
test_img_live['label'] = np.ones(len(test_img_live.index),dtype=int)
```

```python
test_img_live.head()
```

```python
test_img=test_img.append(test_img_live,ignore_index = True)
```

```python
test_img.head()
```

```python
train_img=pd.DataFrame([])
for img in glob.glob("Spoof_data/Training Biometrika Spoof/Training Biometrika Spoof/train spoof/*.png"):
    cv_img = cv2.imread(img)
    newsize = (30, 30)
    img = cv2.resize(cv_img,newsize,interpolation = cv2.INTER_AREA)
    arr = np.array(img)
    one_vector = arr.ravel()
    one_vector=[one_vector]
    image = pd.DataFrame(one_vector)
    train_img=train_img.append(image,ignore_index = True)
```

```python
train_img.head()
```

```python
train_img['label'] = np.zeros(len(train_img.index),dtype=int)
```

```python
train_img.head()
```

```python
train_img_live=pd.DataFrame([])
for img in glob.glob("Spoof_data/Training Biometrika Live/train live/*.png"):
    cv_img = cv2.imread(img)
    newsize = (30, 30)
    img = cv2.resize(cv_img,newsize,interpolation = cv2.INTER_AREA)
    arr = np.array(img)
    one_vector = arr.ravel()
    one_vector=[one_vector]
    image = pd.DataFrame(one_vector)
    train_img_live=train_img_live.append(image,ignore_index = True)
```

```python
train_img_live.head()
```

```python
train_img_live['label'] = np.ones(len(train_img_live.index),dtype=int)
```

```python
train_img_live.head()
```

```python
train_img=train_img.append(train_img_live,ignore_index = True)
```

```python
train_img.head()
```

```python
Y_train=train_img.iloc[:,-1:]
```

```python
X_train=train_img.iloc[:,:-1]
```

```python
Y_test=test_img.iloc[:,-1:]
```

```python
X_test=test_img.iloc[:,:-1]
```

```python
svc_model = SVC(C=0.1,degree=1,kernel='linear')
```

```python
svc_model.fit(X_train,Y_train)
```

```python
Y_pred=svc_model.predict(X_test)
```

```python
print(accuracy_score(Y_test,Y_pred))
```

```python
print(confusion_matrix(Y_test,Y_pred))
```

```python
print(classification_report(Y_test,Y_pred))
```
